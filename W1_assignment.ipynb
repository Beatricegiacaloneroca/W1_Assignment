# Celebrity Scandals and Brand Endorsements: An Analytical Study

In the realm of marketing, celebrity endorsements have long been a cornerstone of brand promotion. However, a pivotal question emerges: ***Can celebrity scandals, despite their potential to tarnish reputations, serve as an efficacious, long-term marketing strategy for the products or brands they represent?***

To address this intriguing inquiry, we turn to a comprehensive analysis of Google Trends data. Through a meticulous examination of three distinctive case studies - Shakira and Piqué, Johnny Depp and Amber Heard, and Georgina Rodríguez and Cristiano Ronaldo - we aim to **study the relationship between celebrity scandals and brand endorsements.**



**Two** of the cases include a scandal:
*   Shakira and Piqué
*   Johnny Depp and Amber Heard

Here the aim is to observe the consequences of the scandal on the interest in both the celebrities and the products they sell.

**One** case, the Georgina Rodríguez and Cristiano Ronaldo, serves as a case to **compare** a product launch and its subsequent interest where there is **NOT**  a previous scandal.



### Step 1: Data Collection

#### Google Trends Data

The first step in our data analysis journey was to collect relevant data from Google Trends. Google Trends is a powerful tool for exploring the popularity and trends associated with specific search queries over time. To investigate the impact of celebrity scandals on brand endorsements, we selected three distinct cases:

1. **Shakira and Piqué**: We collected data related to the search interest for "Shakira," "Piqué," "shakira bzrp," and "Kings League" over the past five years (from today's date).

2. **Johnny Depp and Amber Heard**: For this case, we gathered data for "Johnny Depp" and "Amber Heard" over the same five-year period.

3. **Georgina Rodríguez and Cristiano Ronaldo**: In this case, we examined the search interest for "Georgina," "Ronaldo," and "I am Georgina" over the past five years.

By focusing on these three cases, we aimed to compare scenarios with and without celebrity scandals and assess their impact on search interest. The Google Trends data provides us with insights into how the public's curiosity and engagement change over time, allowing us to draw meaningful conclusions about the effect of scandals on brand endorsements.




```python
# Install pytrends to extract with google trends data
!pip install pytrends

# Import
import pandas as pd
from pytrends.request import TrendReq

# Initialize Google Trends API
pytrends = TrendReq(hl='en-US', tz=360)
```

###Step 2: Data Preprocessing

For each of the three cases, we began by collecting Google Trends data for **specific keywords related to the celebrities and products involved**. In this example (Shakira and Piqué), we focused on terms like 'Shakira,' 'Piqué,' 'shakira bzrp,' and 'Kings League' **over the past 5 years.**

Additionally, a crucial aspect of data preprocessing **was defining the scandal start date**, which, in this case, was set as '2023-01-10.' It's worth noting that this data collection and date definition process was consistently applied to all three cases to ensure uniformity and **facilitate meaningful comparisons in our analysis.**


###Step 3: Exploratory Data Analysis (EDA)

In this step, we conducted Exploratory Data Analysis (EDA) to gain deeper insights into the collected Google Trends data. This involved analyzing search interest trends related to our keywords, which include 'Shakira,' 'Piqué,' 'shakira bzrp,' and 'Kings League' over the past 5 years.

Specifically, for the case of Shakira and Piqué, we focused on:

* Identifying trends and patterns in search interest.
* Visualizing the search interest data over time.
* Identifying any seasonality or recurring patterns.


We also **calculated and assessed the mean search interest** before and after the scandal (defined by the scandal start date of '2023-01-10') during a 15-month observation period. **This allowed us to quantify the impact of the scandal on search interest,** as evidenced by the calculated percentage change.



```python
# Define keywords and timeframes
keywords1 = ['Shakira', 'Piqué', 'shakira bzrp', 'Kings League']
timeframe1 = 'today 5-y'  # Data for the past 5 years

# Build payload and get the interest over time data
pytrends.build_payload(kw_list=keywords1, cat=0, timeframe=timeframe1, geo='', gprop='')

# Get the interest over time data
interest_over_time_df1 = pytrends.interest_over_time()

# Define the scandal start date
scandal_start_date1 = '2023-01-10'

# Define the observation period (15 months before and after the scandal)
observation_period_start1 = pd.to_datetime(scandal_start_date1) - pd.DateOffset(months=15)
observation_period_end1 = pd.to_datetime(scandal_start_date1) + pd.DateOffset(months=15)

# Calculate mean search interest before and after the scandal for the specified observation period
before_scandal_mean1 = interest_over_time_df1[keywords1].loc[observation_period_start1:scandal_start_date1].mean()
after_scandal_mean1 = interest_over_time_df1[keywords1].loc[scandal_start_date1:observation_period_end1].mean()

# Calculate percentage change in search interest
percentage_change1 = ((after_scandal_mean1 - before_scandal_mean1) / before_scandal_mean1) * 100

# Print the results
print(f"Mean Search Interest Before Scandal (15 months): {before_scandal_mean1}")
print(f"Mean Search Interest After Scandal (15 months): {after_scandal_mean1}")
print(f"Percentage Change in Search Interest (15 months): {percentage_change1}")



# ---------------------------------------------------------------------------------
print('\n\n\n\n Part 2: Heard and Depp')

# COMPARISON WITH A. HEARD TRIAL

# Define keywords and timeframes
# keywords = ['Johnny Depp', 'johnny depp movies', 'Amber Heard','heard movies']
keywords2 = ['Johnny Depp', 'Amber Heard'] # changed it to only 2 as there are no other keywords with similar significance
timeframe2 = 'today 5-y'  # Data for the past 5 years


# Build payload and get the interest over time data
pytrends.build_payload(kw_list=keywords2, cat=0, timeframe=timeframe2, geo='', gprop='')

# Get the interest over time data
interest_over_time_df2 = pytrends.interest_over_time()

# Define the scandal start date
scandal_start_date2 = '2022-04-11'  # You can adjust this date based on your analysis needs

# Define the observation period (15 months before and after the scandal)
observation_period_start2 = pd.to_datetime(scandal_start_date2) - pd.DateOffset(months=15) # date of scandal and 15 months before
observation_period_end2 = pd.to_datetime(scandal_start_date2) + pd.DateOffset(months=15) # date of scandal and 15 months after

# Calculate mean search interest before and after the scandal for the specified observation period
before_scandal_mean2 = interest_over_time_df2[keywords2].loc[observation_period_start2:scandal_start_date2].mean()
after_scandal_mean2 = interest_over_time_df2[keywords2].loc[scandal_start_date2:observation_period_end2].mean()

# Calculate percentage change in search interest
percentage_change2 = ((after_scandal_mean2 - before_scandal_mean2) / before_scandal_mean2) * 100

# Print the results
print(f"Mean Search Interest Before Scandal (15 months):{before_scandal_mean2}")
print(f"Mean Search Interest After Scandal (15 months):{after_scandal_mean2}")
print(f"Percentage Change in Search Interest (15 months):{percentage_change2}")



# ---------------------------------------------------------------------------------

# Compare with a case where there is no scandal - Georgina and Ronaldo
print('\n\n\n\n part 3: Georgina and Rondaldo')

# Define keywords and timeframes
keywords3 = ['Georgina', 'Ronaldo', 'I am Georgina']
timeframe = 'today 5-y'  # Data for the past 5 years (adjust as needed)

# Build payload and get the interest over time data
pytrends.build_payload(kw_list=keywords3, cat=0, timeframe=timeframe, geo='', gprop='')

# Get the interest over time data
interest_over_time_df3 = pytrends.interest_over_time()

# Print the data
print(interest_over_time_df3)

# Define the documentary release date
documentary_release_date3 = '2022-01-27'

# Define the observation period (before and after the release)
observation_period_start3 = pd.to_datetime(documentary_release_date3) - pd.DateOffset(months=15)
observation_period_end3 = pd.to_datetime(documentary_release_date3) + pd.DateOffset(months=15)

# Calculate mean search interest before and after the release for the specified observation period
before_release_mean3 = interest_over_time_df3[keywords3].loc[observation_period_start3:documentary_release_date3].mean()
after_release_mean3 = interest_over_time_df3[keywords3].loc[documentary_release_date3:observation_period_end3].mean()

# Calculate percentage change in search interest
percentage_change3 = ((after_release_mean3 - before_release_mean3) / before_release_mean3) * 100

# Print the results
print(f"Mean Search Interest Before Documentary Release (15 months): {before_release_mean3}")
print(f"Mean Search Interest After Documentary Release (15 months): {after_release_mean3}")
print(f"Percentage Change in Search Interest (15 months): {percentage_change3}")

```


###Step 4: Data Visualization

In this step, we harnessed data visualization techniques to gain a deeper understanding of the search interest trends for the three cases:

* Shakira and Piqué: We observed the percentage change in search interest for keywords related to Shakira, Piqué, 'shakira bzrp,' and 'Kings League.' The bar chart visualizations provide insights into the impact of the scandal that began on '2023-01-10' on the search interest surrounding these keywords. Each bar represents the percentage change in search interest, and we conducted this analysis over a 15-month period before and after the scandal.

* Johnny Depp & Amber Heard Trial: Similarly, we visualized the percentage change in search interest for keywords related to Johnny Depp and Amber Heard, focusing on the trial that commenced on '2022-04-11.' These bar charts help us understand how the trial affected search interest over a 15-month observation period.

* Georgina & Ronaldo Documentary: In this case, we investigated the impact of a documentary release on the search interest for keywords like 'Georgina,' 'Ronaldo,' and 'I am Georgina.' The visualizations depict the percentage change in search interest over a 15-month period, starting from the documentary's release date on '2022-01-27.'

These visualizations allow us to compare and contrast the consequences of scandals and product launches on search interest, helping us answer our data science question effectively.

```python
# Data visualization

import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt


# Initialize Google Trends API
pytrends = TrendReq(hl='en-US', tz=360)

# Define keywords and timeframes for each case
cases = [
    {
        'name': 'Shakira',
        'keywords': ['Shakira', 'Piqué', 'shakira bzrp', 'Kings League'],
        'start_date': '2023-01-10',
        'months_before': 15,
        'months_after': 15
    },
    {
        'name': 'Johnny Depp & Amber Heard Trial',
        'keywords': ['Johnny Depp', 'Amber Heard'],
        'start_date': '2022-04-11',
        'months_before': 15,
        'months_after': 15
    },
    {
        'name': 'Georgina & Ronaldo Documentary',
        'keywords': ['Georgina', 'Ronaldo', 'I am Georgina'],
        'start_date': '2022-01-27',
        'months_before': 15,
        'months_after': 15
    }
]

# Function to calculate percentage change
def calculate_percentage_change(case):
    pytrends.build_payload(kw_list=case['keywords'], cat=0, timeframe='today 5-y', geo='', gprop='')
    interest_over_time_df = pytrends.interest_over_time()

    observation_period_start = pd.to_datetime(case['start_date']) - pd.DateOffset(months=case['months_before'])
    observation_period_end = pd.to_datetime(case['start_date']) + pd.DateOffset(months=case['months_after'])

    before_mean = interest_over_time_df[case['keywords']].loc[observation_period_start:case['start_date']].mean()
    after_mean = interest_over_time_df[case['keywords']].loc[case['start_date']:observation_period_end].mean()

    percentage_change = ((after_mean - before_mean) / before_mean) * 100
    return percentage_change

# Calculate and plot percentage change for each case
for case in cases:
    percentage_change = calculate_percentage_change(case)
    plt.figure(figsize=(10, 5))
    plt.title(f'Percentage Change in Search Interest for {case["name"]}', fontsize=16)
    percentage_change.plot(kind='bar', color='skyblue')
    plt.xlabel('Keywords')
    plt.ylabel('Percentage Change')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
